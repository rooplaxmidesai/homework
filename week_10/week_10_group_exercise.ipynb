```python
#Function to calculate dot product of vectors
def dot_prod(v1,v2):
        return sum(i*j for (i,j) in zip(v1, v2))
    
```

### Cross Product
##### Example: The cross product of a = (2,3,4) and b = (5,6,7)
* cx = aybz − azby = 3×7 − 4×6 = −3
* cy = azbx − axbz = 4×5 − 2×7 = 6
* cz = axby − aybx = 2×6 − 3×5 = −3
##### Answer: a × b = (−3,6,−3)

```python
#Call numpy cross product and cross_prod function. Including try and except for error handling

try:
    
    #Using numpy cross product
    print('Using numpy cross product')
    print(np.cross(np.array([5,-4]), np.array([-7,2]), axisa=0, axisb=0))
    print(np.cross(np.array([5,-4,3]), np.array([-7,2,-8]), axisa=0, axisb=0))
    print(np.cross(np.array([2,3,4]), np.array([5,6,7]), axisa=0, axisb=0))
    print(np.cross(np.array([3,5,-7]), np.array([2,-6,4]), axisa=0, axisb=0))
    print(np.cross(north,east))
    print(np.cross(west,east))
    
    #Using function cross_prod
    print('Using function cross_prod')
    print(cross_prod(np.array([5,-4]),np.array([-7,2])))
    print(cross_prod(np.array([5,-4,3]),np.array([-7,2,-8])))
    print(cross_prod(np.array([2,3,4]),np.array([5,6,7])))
    print(cross_prod(np.array([3,5,-7]),np.array([2,-6,4])))
    print(cross_prod(north,east))
    print(cross_prod(west,east))

except:
    print("Not valid parameters")
```

```python
#Function to calculate cross product of vectors
def cross_prod(v1,v2):
    result = 0
    x=0
    y=1
    z=2
    if v1.shape[0] == 3 :
        cx = (v1[y]*v2[z]) - (v1[z]*v2[y])
        cy = (v1[z]*v2[x]) - (v1[x]*v2[z])
        cz = (v1[x]*v2[y]) - (v1[y]*v2[x])

        result = np.array([cx,cy,cz])
        
    if v1.shape[0] == 2:
        cx = (v1[x]*v2[y]) - (v1[y]*v2[x])
        result = cx
        
    return result
```

## 2. What is one hot encoding?
* One hot encoding is representing categorical values into binary values(0/1). When using machine algorithms, the categorical values(which are string) can not be processed and hence needs to be converted to numerical values. If we just assign a numerical value like 1,2,3.. then machine can interprete it as one value is greater or less than the other. Hence, it needs to be one hot encoded to be processed by algorithm.


```python
belize_airbnb_df = pd.read_csv("belize_airbnb_listings.csv")
belize_airbnb_df.head(1)
```

```python
#One hot encoding for neighbourhood_group column of dataframe
# Using .loc
belize_hotencoded_df = belize_airbnb_df

belize_hotencoded_df['Belize'] = 0
belize_hotencoded_df['Stann Creek'] = 0
belize_hotencoded_df['Cayo'] = 0
belize_hotencoded_df['Toledo'] = 0
belize_hotencoded_df['Corozal'] = 0
belize_hotencoded_df['Orange Walk'] = 0

belize_hotencoded_df.loc[belize_hotencoded_df["neighbourhood_group"]=="Belize","Belize"] = 1
belize_hotencoded_df.loc[belize_hotencoded_df["neighbourhood_group"]=="Stann Creek","Stann Creek"] = 1
belize_hotencoded_df.loc[belize_hotencoded_df["neighbourhood_group"]=="Cayo","Cayo"] = 1
belize_hotencoded_df.loc[belize_hotencoded_df["neighbourhood_group"]=="Toledo","Toledo"] = 1
belize_hotencoded_df.loc[belize_hotencoded_df["neighbourhood_group"]=="Corozal","Corozal"] = 1
belize_hotencoded_df.loc[belize_hotencoded_df["neighbourhood_group"]=="Orange Walk","Orange Walk"] = 1

belize_hotencoded_df
```

```python
#One hot encoding for neighbourhood_group column of dataframe
#Using dummies
dummies = pd.get_dummies(belize_airbnb_df.neighbourhood_group)
dummies.head(5)
```

```python
merged_hotencoded = pd.concat([belize_airbnb_df,dummies],axis='columns')
merged_hotencoded.head(5)
```
